```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
df1 = pd.read_csv('world-happiness-report.csv')
df2 = pd.read_csv('world-happiness-report-2021.csv')
hasil=df2.drop(columns=['Standard error of ladder score','upperwhisker','lowerwhisker','Explained by: Log GDP per capita','Ladder score in Dystopia','Explained by: Log GDP per capita','Explained by: Social support','Explained by: Healthy life expectancy','Explained by: Freedom to make life choices','Explained by: Generosity','Explained by: Perceptions of corruption'])

hasil.describe()

```

```python
Soal 3: Happiness report negara Indonesia dari tahun 2006 - 2021
Output yang diharapkan adalah data negara Indonesia dari penggabungan dataset happiness report dan dataset happiness report 2021. Untuk sesuai dengan expected ouput, lakukan perintah-perintah berikut:

Lakukan filtering dataset happiness report dari kolom 'Country name'== 'Indonesia' - tambahkan atribut .reset_index(drop=True) - drop kolom 'Positive affect' dan 'Negative affect'

Lakukan filtering dataset dari variabel no 2 hasil dropping beberapa kolom - tambahkan kolom 'year'=2021 pada dataset dari variabel no 2 hasil dropping beberapa kolom - filtering dataset tersebut menggunakan kolom 'Country name'== 'Indonesia' - rename beberapa kolom dengan menggunakan atribut .rename() dari pandas, referensi dokumentasi : https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html {
'Ladder score':'Life Ladder', 'Logged GDP per capita':'Log GDP per capita', 'Healthy life expectancy':'Healthy life expectancy at birth'}

    - drop kolom 'Dystopia + residual' dan 'Regional indicator'
Gabungkan data dari hasil poin 1 dan poin 2

gunakan pandas concat, .concat()
tambahkan atribut .reset_index(drop=True)
rename kolom {'year':'Year'}
```

```python
indonesia=df1['Country name']=='Indonesia'
hasil2=df1[indonesia].reset_index(drop=True)
hasil_akhir=hasil2.drop(columns=['Positive affect','Negative affect'])
# display(hasil_akhir)
hasil['year']=2021
filter_indonesia=hasil['Country name']=='Indonesia'
data_2=hasil[filter_indonesia].reset_index(drop=True)
hasil_drop=data_2.drop(columns=['Dystopia + residual','Regional indicator'])
data_rename=hasil_drop.rename(columns={"Ladder score":"Life Ladder","Logged GDP per capita":"Log GDP per capita","Healthy life expectancy":"Healthy life expectancy at birth"})
gabungan_data=pd.concat([hasil_akhir,data_rename])
hasil_akhir_finish=gabungan_data.rename(columns={'year':'Year'}).reset_index(drop=True)
display(hasil_akhir_finish)


```

```python
# Soal 4: Visualisasi lineplot Life Ladder negara Indonesia
# Gunakan library seaborn untuk plotting line plot kolom Life Ladder dari hasil penggabungan negara Indonesia tahun 2006-2021 (soal 3) seperti expected output, jika tidak bisa menyelesaikan soal 3 silahkan download data berikut Download.csv
# Berikan insight dari hasil visualisasi tersebut

df_drop=hasil_akhir_finish.drop(columns=['Country name','Log GDP per capita','Social support','Healthy life expectancy at birth','Freedom to make life choices','Generosity','Perceptions of corruption'])
sns.set({'figure.figsize':(12,8)})
sns.lineplot(x='Year',y='Life Ladder', data=df_drop, marker='o')
plt.title('Indonesia ladder score over the year')
plt.show()
```

```python
jawaban inshight dari hasil visualisasi:
berdasarkan grafik Indonesia ladder score over the year diatas dapat di simpulkan sebagai berikut:
1.life ladder tertinggi di dapatkan pada tahun 2014
2. sedangkan life ladder terendah di dapatkan pada tahun 2006
3. nilai life ladder nya bersifat fluktuatif berdasarkan nilai grafik diatas
```

```python
Soal 6: Visualisasi Scatter plot
Buatlah visualisasi scatter plot (perpaduan seaborn dan matplotlib) dari df2 antara GDP per capita ('Logged GDP per capita') dan skor korupsi ('Perceptions of corruptions') sesuai expected output dengan ketentuan berikut:
Gunakan kolom skor kebebasan beraksi ('Freedom to make life choices') sebagai ukuran pointnya kalikan suatu angka yang cukup besar agar terlihat.
Gunakan kolom skor kesehatan ('Healthy life expectancy') sebagai warna point, berikan cmap 'plasma'.
Gunakan regplot pada seaborn (sns.regplot) dengan argumen scatter=False dan color = 'black'
df2
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x=df2['Logged GDP per capita'],y=df2['Perceptions of corruption'],c=df2['Healthy life expectancy'],cmap='plasma',s=df2['Freedom to make life choices']*400,alpha=0.6)
sns.regplot(x=df2['Logged GDP per capita'],y=df2['Perceptions of corruption'],scatter=False,color='black')
plt.colorbar()
plt.title('Coruption vs GDP per capita ')

plt.show()
```

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 

df=sns.load_dataset('penguins')
df 
```

```python
df.info()
```

```python

Soal 2. Persiapan Data
Hapus nilai missing dalam rows dataset menggunakan atribut.dropna()
Lakukan grouping data berdasarkan kolom 'species' dengan menggunakan agregasi mean()

```

```python
df.dropna(axis=0, inplace=True)
df
```

```python
df.info()
```

```python
df_group=df.groupby(df['species']).mean()
df_group
```

```python
Soal 3. Univariate Analysis
Buatlah sebuah analisa univariate dengan menggunakan perintah seperti berikut:

Buatlah sebuah Pie plot dari dataset hasil grouping untuk kolom flipper_length_mm seperti expected output berikut:
Buatlah sebuah plot histogram dari dataset awal (sebelum grouping) untuk kolom body_mass_g
Jelaskan insight yang diperoleh dari ke dua grafik tersebut.
```

```python
plt.figure(figsize=(12,8))
plt.pie(df_group['flipper_length_mm'], labels=df_group.index,autopct='%.1f%%')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.hist(df['body_mass_g'],bins=10)
plt.xlabel('body_mass_g')
plt.ylabel('count')
plt.show()
```

```python
-penjelasan dari grafik di atas sebagai berikut :
-untuk grafik pada diagram pie di dapat insight sbb:
1. pada diagram pie diatas nilai tertinggi diproleh oleh Gentoo sedangkan nilai terendah di proleh oleh adelie
2. nilai tertinggi di tandai dengan warna Hijau sedangkan nilai terendah ditandai dengan warna biru 

  -untuk grafik pada diagram hist di dapat insight sbb:
1. pada diagram histogram nilai berat (body_mass_g) dianatara 3500 memiliki jumlah yang yang paling tinggi
 sedangkan yang nialai terendah di peroleh diantara berat (body_mass_g) 6000
 
 

```

```python
Soal 4. Bivariate Analysis
Buatlah sebuah grafik catplot menggunakan seaborn dengan x='sex', y='bill_length_mm' dan kind 'bar' seperti expected output di bawah
Buatlah sebuah grafik scatter plot menggunakan seaborn dengan x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm' seperti expected dibawah
Jelaskan insight yang diperoleh dari hasil analisa bivariate tersebut
```

```python
plt.figure(figsize=(12,8))
sns.catplot(x='sex', y='bill_length_mm', kind='bar', data=df)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='body_mass_g',y='flipper_length_mm',hue='bill_length_mm',data=df)
plt.show()
```

```python
penjelasan insight yang diproleh dari kedua grafik di atas :
-untuk grafik pada catplot
   pada grafik diatas di dapatkan nilai (bill_length_mm) jenis kelamin untuk Male lebih  tinggi dari pada nilai (bill_length_mm)  jenis    kelamin untuk  Female
 
-untuk grafik scatterplot

```

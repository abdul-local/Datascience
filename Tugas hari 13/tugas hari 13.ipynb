```python
# Soal 1: Multiple Subplots
# Dengan data di bawah ini buatlah visualisasi seperti expected output:

import matplotlib.pyplot as plt 
import numpy as np
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(12,8),sharex=True, sharey=True)
ax.flatten()
ax[0].plot(x,tan,color='blue',label='tanx')
ax[1].plot(x,sin,color='red',label='sinx')
ax[2].plot(x,cos,color='green',label='cosx')
ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')
plt.xlim(-6,6)
plt.ylim(-10,10)
fig.legend(loc='upper center')

plt.show()
```

```python
# Soal 2: Nested Axis
# Dengan data di bawah ini, buatlah visualisasi seperti expected output yang diinginkan:
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2= np.cos(x**2)
y3= np.cos(x**3)
y4= np.cos(x**4)
y5= np.cos(x**5)
fig,ax=plt.subplots(figsize=(12,8),sharex=True,sharey=True)
ax.plot(x,y)
ins1=ax.inset_axes([0.9,0.9,0.3,0.3])
ins1.set_xlim(xmin=-6,xmax=6)
ins1.plot(x,y4,color='black')
ins1.plot(x,y,color='blue')
ins1.set_title('Grafik cos^4')
ins2=ax.inset_axes([-0.1,0.9,0.3,0.3])
ins2.set_xlim(xmin=-6,xmax=6)
ins2.plot(x,y5,color='yellow')
ins2.set_title('Grafik cos^5')
ins2.plot(x,y,color='blue')
ins3=ax.inset_axes([-0.1,-0.0,0.3,0.3])
ins3.set_xlim(xmin=-6,xmax=6)
ins3.plot(x,y2,color='red')
ins3.set_title('Grafik cos^2')
ins3.plot(x,y,color='blue')
ins4=ax.inset_axes([0.9,0.0,0.3,0.3])
ins4.set_xlim(xmin=-6,xmax=6)
ins4.plot(x,y3,color='green')
ins4.set_title('Grafik cos^3')
ins4.plot(x,y,color='blue')
ax.set_title('Grafik cosx')
plt.show()
```

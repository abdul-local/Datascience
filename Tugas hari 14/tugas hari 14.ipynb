```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('vgsales.csv')
df.head()
```

```python
# <h2>Soal 1: Preparing Data</h2>
# Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

dfgroupby=df.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].mean()
dfgroupby
```

```python
# <h2>Soal 2: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>
# 1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
# 2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

import numpy as np
plt.figure(figsize=(12,8))
x=np.arange(4)
width=0.07
for i,j in enumerate(dfgroupby.T.columns):
    # print(i,j)
    # print(i)
    plt.bar(x+(width*i),dfgroupby.T[j],width,label=j)
plt.legend(loc='upper right',bbox_to_anchor=(1.17,1.17),framealpha=0.5,facecolor='grey')
plt.title('Mean Sales Video Games By Genre')
plt.ylabel('Mean Sales')
plt.xlabel('Region Sales')
plt.xticks(x+0.2,labels=dfgroupby.T.index)
plt.show()
```

```python
#Berdasarkan grafik Barplot di atas dapat di berikan kesimpulan:
1. pada daerah penjualan Video Game  NA_Sales di dapatkan nilai rata-rata penjualan tertinggi yang di proleh oleh Platform
sedangakan nilai rata-rata penjualan terendah diproleh oleh Adventure
2. pada daerah penjualan Video Game  EU_Sales di dapatkan nilai rata-rata penjualan tertinggi yang di proleh oleh Shooter
sedangakan nilai rata-rata penjualan terendah diproleh oleh Adventure
3. pada daerah penjualan Video Game  JP_Sales di dapatkan nilai rata-rata penjualan tertinggi yang di proleh oleh Role-Playing
sedangakan nilai rata-rata penjualan terendah diproleh oleh Shooter
4. pada daerah penjualan Video Game  Other_Sales di dapatkan nilai rata-rata penjualan tertinggi yang di proleh oleh Shooter
sedangakan nilai rata-rata penjualan terendah diproleh oleh Advanture
5. dari hasil kesimpulan di point 1 sampai 4 terdapat nilai rata-rata penjualan tertinggi berbeda-beda tergantung daerah penjualannya dan untuk daerah penjualan tertinggi dari grafik di proleh rata-rata pada daerah NA_Sales dan rata-rata terendah diproleh oleh daerah/region Other_Sales

```

```python
# <h2>Soal 3: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

# 1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
# 2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut
plt.figure(figsize=(12,8))
x=np.arange(4)
width=0.7
y_bottom=0
for i,j in enumerate(dfgroupby.T.columns):
    # print(j)
    plt.bar(x,dfgroupby.T[j],width,y_bottom,label=j)
    y_bottom+=dfgroupby.T[j]
plt.legend(loc='upper right',bbox_to_anchor=(1.17,1.17),framealpha=0.5,facecolor='grey')
plt.title('Mean Sales Video Games By Genre')
plt.ylabel('Mean Sales')
plt.xlabel('Region Sales')
plt.xticks(x+0.02,labels=dfgroupby.T.index)
plt.show()
```
